Usando o dataset seeds.csv, encontre o melhor número de clusters para o K-means usando *inertia* e *silhouette coefficient*

```python
import pandas as pd

df = pd.read_csv('seeds.csv')
# del df['grain_variety']
df.head()
```

```python
X = df.drop(columns=['grain_variety']).values
y = df['grain_variety']
```

### TO-DO

* Crie uma instância do K-means chamada *model* com k clusters (k variando de 1 a 6)
* Treine um modelo usando os dados (X) para cada k
* adicione o valor de *inertias_* de cada modelo numa lista chamada inertias

```python
from sklearn.cluster import KMeans

inertias = []

# Train models for each k and append inertias
for k in range(1, 7):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertias.append(model.inertia_)
    
for k, inertia in enumerate(inertias, start=1):
    print(f"K={k}: Inertia={inertia}")
```

### TO-DO
* plote as inércias para determinar o melhor valor de k

```python
import matplotlib.pyplot as plt

# resposta# Plot the inertias to determine the best value of k
plt.figure(figsize=(10, 6))
plt.plot(range(1, 7), inertias, marker='o', linestyle='--', color='b')
plt.title('Inertia for each k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### TO-DO
* Encontre o melhor valor de k usando *silhouette coefficient* 


```python
from sklearn.metrics import silhouette_score
import numpy as np

# Calculate silhouette scores for different k values
silhouette_scores = []
k_range = range(2, 7)  # Start from 2 as silhouette score is not defined for k=1

for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    cluster_labels = model.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For k={k}, the average silhouette score is: {silhouette_avg:.4f}")

# Find the best k (highest silhouette score)
best_k = k_range[np.argmax(silhouette_scores)]
print(f"\nThe best k value based on silhouette score is: {best_k}")

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o', linestyle='--', color='b')
plt.title('Silhouette Score for each k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```
